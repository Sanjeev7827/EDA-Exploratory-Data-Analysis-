```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r"D:\Datasets(Github)\Restaurant Datset.csv")
```

```python
df
```

 # Q.1. What was the Most Preferred Payment Method ?

```python
df['Payment Method'].value_counts()
```

```python
sns.countplot(x='Payment Method',data=df,palette='inferno')
plt.legend(['Cash','Credit Card','Gift Card'])
plt.xlabel('Payment Method',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

 # Q.2. Which one was the Most Selling Product 
 # a.By Quantity
 # b.By Revenue

```python
df.head()
```

```python
df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
df.groupby('Product')['Quantity'].sum().plot(kind='bar',color='r')
plt.xticks(rotation=50)
plt.legend()
plt.title('Most Selling Product by Quantity',fontsize=15)
plt.xlabel('Products',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
df['Revenue']=df['Price']*df['Quantity']
```

```python
df
```

```python
df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
```

```python
df.groupby('Product')['Revenue'].sum().plot(kind='bar',color='c')
plt.xticks(rotation=50)
plt.legend()
plt.title('Most Selling Product by Revenue',fontsize=15)
plt.xlabel('Products',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

 # Q.3. Which City had maximum revenue, or, Which Manager earned maximum revenue ?


```python
df.head()
```

```python
df.groupby('City')['Revenue'].sum().sort_values(ascending=False)
```

```python
df.groupby('City')['Revenue'].sum().plot(kind='bar',color='y')
plt.xticks(rotation=50)
plt.legend()
plt.title('Most Revenue Generate by City',fontsize=15)
plt.xlabel('City',fontsize=15)
plt.ylabel('Count(US $)',fontsize=15)
plt.show()
```

```python
df.groupby('Manager')['Revenue'].sum().sort_values(ascending=False)
```

```python
df.groupby('Manager')['Revenue'].sum().plot(kind='bar',color='m')
plt.xticks(rotation=50)
plt.legend()
plt.title('Most Revenue Generate by City',fontsize=15)
plt.xlabel('Manager',fontsize=15)
plt.ylabel('Count(US $)',fontsize=15)
plt.show()
```
